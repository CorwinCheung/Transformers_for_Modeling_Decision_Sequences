```python
import torch
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Code for actions:
*   0 == l
*   1 == r
*   2 == L
*   3 == R

choice = 0 or 1 meaning left or right <br>
rewarded = 0 or 1 meaning yes or no <br>
then then code = choice + 2*rewarded

```python
# initialize model parameters
alpha = .5
beta = 2
tau = 1.2
pHi=0.8
pLow=1-pHi
stateChangeProb=0.02

# generator
g=torch.Generator().manual_seed(1999)
np_g=np.random.default_rng(1999)
```

```python
#run model - RFLR from Beron et al
modelRuns=500000

# initialize storage
sumWords=np.zeros(modelRuns, dtype='int')

# initialize model
phi=0.0
LOR=0.0

# randomize some initial conditions
hiPort=0 if torch.rand(1, generator=g).item()>0.5 else 1
lastChoice=0 if torch.rand(1, generator=g).item()>0.5 else 1
```

```python
for counter in range(modelRuns):

  # was is prob of choosing left
  LOR=alpha*lastChoice+phi
  pRight=math.exp(LOR)/(1+math.exp(LOR))

  # does it come up left ?
  choice=1 if torch.rand(1, generator=g).item()<=pRight else 0

  # is it rewarded, based on choice, hiPort, pHi, and pLow
  if choice==hiPort:
    rewarded=1 if torch.rand(1, generator=g).item()<=pHi else 0
  else:
    rewarded=1 if torch.rand(1, generator=g).item()<=pLow else 0

  code=choice+2*rewarded
  phi = beta*(2*choice-1)*rewarded + math.exp(-1/tau)*phi

  sumWords[counter]=code

  lastChoice=choice

  # do we switch ports?
  # prevent switches in the last 20 trials as that causes problems when averaging
  # across block switches
  if (counter<modelRuns-20) and torch.rand(1, generator=g).item()<=stateChangeProb:
    print(counter, 'switch')
    hiPort=1-hiPort
```
